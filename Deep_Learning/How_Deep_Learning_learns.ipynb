```python
import tensorflow as tf
model = tf.keras.models.Sequential()
```

# 손실 함수(Loss function)

## MSE(Mean Squared Error, MSE)

```python
model.compile(optimizer='adam', loss='mse', metrics=['mse'])
```

```python
model.compile(optimizer='adam', loss=tf.keras.losses.MeanSquaredError(), metrics=['mse'])
```

## 이진 크로스 엔트로피(Binary Cross-Entropy)

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
```

```python
model.compile(loss=tf.keras.losses.BinaryCrossentropy(), optimizer='adam', metrics=['acc'])
```

## 카테고리칼 크로스 엔트로피(Categorical Cross-Entropy)

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['acc'])
```

```python
model.compile(loss=tf.keras.losses.CategoricalCrossentropy(), optimizer='adam', metrics=['acc'])
```

- 레이블에 대해서 원-핫 인코딩 과정을 생략하고, 정수값을 가진 레이블에 대해서 다중 클래스 분류를 수행하고 싶다면 다음과 같이 'sparse_categorical_crossentropy'를 사용합니다.

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['acc'])
```

```python
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(), optimizer='adam', metrics=['acc'])
```

그 외 다양한 손실 함수들

# 옵티마이저(Optimizer)

## 모멘텀(Momentum)

```python
tf.keras.optimizers.SGD(learning_rate=0.01, momentum=0.9)
```

## 아다그라드(Adagrad)

```python
tf.keras.optimizers.Adagrad(learning_rate=0.01, epsilon=1e-6)
```

## 알엠에스프롭(RMSprop)

```python
tf.keras.optimizers.RMSprop(learning_rate=0.001, rho=0.9, epsilon=1e-06)
```

## 아담(Adam)

```python
tf.keras.optimizers.Adam(learning_rate=0.0001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
```

## 사용 방법

```python
adam = tf.keras.optimizers.Adam(learning_rate=0.0001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['acc'])
```
