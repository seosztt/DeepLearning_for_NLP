# 정규 표현식 실습

```python
import re
```

## . 기호

```python
r = re.compile("a.c")
r.search('kkk')
```

```python
r.search('abc')
```

## ? 기호

```python
r = re.compile('ab?c')
r.search('abbc')
```

```python
r.search('abc')
```

```python
r.search('ac')
```

## * 기호

```python
r = re.compile('ab*c')
r.search('a')
```

```python
r.search('ac')
```

```python
r.search('abc')
```

```python
r.search('abbbbc')
```

## + 기호

```python
r = re.compile('ab+c')
r.search('ac')
```

```python
r.search('abc')
```

```python
r.search('abbbbc')
```

## ^ 기호

```python
r = re.compile('^ab')

r.search('bbc')
r.search('zab')
```

```python
r.search('abz')
```

## {숫자} 기호

```python
r = re.compile('ab{2}c')

r.search('ac')
r.search('abc')
r.search('abbbbbbc')
```

```python
r.search('abbc')
```

## {숫자1, 숫자2} 기호

```python
r = re.compile('ab{2,8}c')

r.search('ac')
r.search('abc')
r.search('abbbbbbbbbbbc')
```

```python
r.search('abbc')
```

```python
r.search('abbbbbbbc')
```

## {숫자,} 기호

```python
r = re.compile('a{2,}bc')

r.search('bc')
r.search('aa')
```

```python
r.search('aabc')
```

```python
r.search('aaaaaaabc')
```

## [ ] 기호

```python
r = re.compile('[abc]') # [abc]는 [a-c]와 같다.
r.search('zzz') # 아무런 결과도 출력되지 않는다.
```

```python
r.search('a')
```

```python
r.search('aaaaaaa')
```

```python
r.search('baac')
```

```python
r = re.compile('[a-z]')

# 아무런 결과도 출력되지 않는다.
r.search('AAA')
r.search('111')
```

```python
r.search('aBC')
```

## [^문자] 기호

```python
r = re.compile('[^abc]')

r.search('a')
r.search('ab')
r.search('b')
```

```python
r.search('d')
```

```python
r.search('1')
```

# 정규 표현식 모듈 함수 예제

## re.match() 와 re.search()의 차이

```python
r = re.compile('ab.')
r.match('kkkabc')
```

```python
r.search('kkkabc')
```

```python
r.match('abckkk')
```

## re.split()

```python
# 공백 기준 분리
text = '사과 딸기 수박 메론 바나나'
re.split(' ', text)
```

```python
# 줄바꿈 기준 분리
text = '''사과
딸기
수박
메론
바나나'''

re.split('\n', text)
```

```python
# '+'를 기분으로 분리
text = "사과+딸기+수박+메론+바나나"

re.split('\+', text)
```

## re.findall()

```python
text = """이름 : 김철수
전화번호 : 010 - 1234 - 1234
나이 : 30
성별 : 남"""

re.findall('\d+', text)
```

```python
re.findall('\d+', '문자열입니다.')
```

## re.sub()

```python
text = "Regular expression : A regular expression, regex or regexp[1] (sometimes called a rational expression)[2][3] is, in theoretical computer science and formal language theory, a sequence of characters that define a search pattern."

preprocessed_text = re.sub('[^a-zA-Z]', ' ', text)
print(preprocessed_text)
```

# 정규 표현식 텍스트 전처리 예제

```python
re.findall('[A-Z]', text)
```

```python
re.findall('[A-Z]{4}', text)
```

# 정규 표현식을 이용한 토큰화

```python
from nltk.tokenize import RegexpTokenizer

text = "Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery goes for a pastry shop"

tokenizer1 = RegexpTokenizer("[\w]+")
tokenizer2 = RegexpTokenizer('\s+', gaps=True)

print(tokenizer1.tokenize(text))
print(tokenizer2.tokenize(text))
```
